```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('BarChart_DataFrame.csv',sep=',',index_col=0)
print(df)
```

```python
fig=plt.figure(figsize=(10,8),dpi=150)
df.plot(kind='bar')
plt.tight_layout()
plt.show()
```

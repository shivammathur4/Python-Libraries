```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
country_df=pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')
```

```python
country_df.info()
```

```python
sns.boxplot(x='continent',y='lifeExp',data=country_df)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
sns.violinplot(x=df.species,y=df.sepal_length)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips_dataFrame=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
```

```python
tips_dataFrame.info()
```

```python
tips_dataFrame.describe()
```

```python
sns.boxplot(x='day',y='tip',data=tips_dataFrame,hue='sex')
plt.show()
```

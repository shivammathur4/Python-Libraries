```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gold_medal=pd.read_csv('GoldMedal.csv')
print(gold_medal)
```

```python
labels=list(gold_medal['country'])
slices_tally=list(gold_medal['gold_medal'])
explode=(0.1,0,0,0,0)
```

```python
plt.pie(slices_tally,labels=labels,wedgeprops={'edgecolor':'black'},explode=explode)
plt.title('Gold Medal Tally')
plt.tight_layout()
plt.show()
```

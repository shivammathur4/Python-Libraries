```python
import matplotlib.pyplot as plt
```

```python
xValues=[1,2,3]
yValues=[2,4,1]
```

```python
plt.plot(xValues,yValues)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Draw a line graph')
plt.show()
```

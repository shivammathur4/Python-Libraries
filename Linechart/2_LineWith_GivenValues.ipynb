```python
import matplotlib.pyplot as plt
```

```python
xValues=[1,2,3,4,5]
yValues=[4,6,7,2,4]
```

```python
plt.plot(xValues,yValues)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('line graph X and Y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xValues = [0,5,10,15]
yValues= [10,20,30,40]
plt.show()
plt.plot(xValues,yValues,label='Line 1',color='skyblue',linewidth=4,marker='o',mfc='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines with current axis limit values')
print(plt.axis((0,100,0,200)))
plt.legend()
plt.show()
```

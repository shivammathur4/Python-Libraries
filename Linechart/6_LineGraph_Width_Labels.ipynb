```python
import matplotlib.pyplot as plt
```

```python
X_line1=[10,25,45,35]
Y_line1=[30,40,20,15]
plt.plot(X_line1,Y_line1,label='Line 1',color='green',linewidth=4)
X_line2=[30,40,15,20]
Y_line2=[20,35,10,45]
plt.plot(X_line2,Y_line2,label='Line 2',color='blue',linewidth=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two lines presented on same plot')
plt.legend()
plt.show()
```

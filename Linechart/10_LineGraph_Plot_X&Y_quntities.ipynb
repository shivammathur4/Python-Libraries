```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1=[1,3,4,5,6]
Y1=[3,2,4,8,2]
X2=[4,6,7,5,2]
Y2=[2,3,6,4,7,]
plt.axis([0, 10, 0, 30]) 
plt.plot(X1, Y1,'b*', X2, Y2, 'ro')
plt.legend()
plt.title('X and Y quantities')
plt.show()
```

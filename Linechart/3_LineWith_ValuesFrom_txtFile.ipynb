```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
valuesFromTxt=np.loadtxt('3_test.txt',dtype=int)
```

```python
X_values=valuesFromTxt[:,0:1].flatten()
Y_values=valuesFromTxt[:,1:].flatten()
```

```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line graph values for txt file')
plt.plot(X_values,Y_values)
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
xvalue = [1, 2, 3]
yvalue = [0, 1, 0]
zvalue = [1, 0, 1]
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(xvalue, yvalue)
ax[1].plot(xvalue, zvalue)
```

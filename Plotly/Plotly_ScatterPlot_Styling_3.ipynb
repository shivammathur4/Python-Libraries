```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.express as px
```

```python
fig=go.Figure(data=go.Scattergl(x=np.random.rand(500),
                                y=np.random.rand(500),
                                mode='markers',
                            marker=dict(color=np.random.rand(10),
                            colorscale='inferno',line_width=1) 
                                ))
```

```python
fig.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('13_CSV.csv', index_col='Date')
```

```python
plt.plot(df)
plt.minorticks_on()
plt.grid(color = 'black', linestyle = '--')
plt.xlabel('Dates')
plt.ylabel('Close')
plt.title('Alphabet INC')
plt.legend(df.columns)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('SeaLevel.csv',sep=',',parse_dates=True)
df.info()
```

```python
Time=df['Time']
y_GMSL=df['GMSL']
plt.scatter(Time,y_GMSL)
plt.xlabel('Year')
plt.ylabel('Rising Sea level')
#plt.xticks(np.arange(1993,2016,2))
plt.show()
```
